```python
# import libraries (you may add additional imports but you may not have to)
import numpy as np
import pandas as pd
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
```

```python
# get data files
#!wget https://cdn.freecodecamp.org/project-data/books/book-crossings.zip

#!unzip book-crossings.zip

books_filename = 'BX-Books.csv'
ratings_filename = 'BX-Book-Ratings.csv'
```

```python
# import csv data into dataframes
df_books = pd.read_csv(
    books_filename,
    encoding = "ISO-8859-1",
    sep=";",
    header=0,
    names=['isbn', 'title', 'author'],
    usecols=['isbn', 'title', 'author'],
    dtype={'isbn': 'str', 'title': 'str', 'author': 'str'})

df_ratings = pd.read_csv(
    ratings_filename,
    encoding = "ISO-8859-1",
    sep=";",
    header=0,
    names=['user', 'isbn', 'rating'],
    usecols=['user', 'isbn', 'rating'],
    dtype={'user': 'int32', 'isbn': 'str', 'rating': 'float32'})
df_books.head()
```

```python
df_ratings.head()
```

```python
import pandasql as ps
query="""SELECT isbn, COUNT() AS total FROM df_ratings

        GROUP BY isbn
        ORDER BY total"""

cleaned_df=ps.sqldf(query, locals())
cleaned_df=cleaned_df[cleaned_df["total"]>=100]
cleaned_df.head()
```

```python

query="""SELECT user, COUNT() AS total FROM df_ratings

        GROUP BY user
        ORDER BY total"""

user_df=ps.sqldf(query, locals())
user_df=user_df[user_df["total"]>=200]
user_df.head()
```

```python

df_ratings = df_ratings.merge(user_df, on = "user", how="inner")
df_ratings = df_ratings.merge(cleaned_df, on = "isbn", how="inner")
df_ratings = df_ratings.merge(df_books, on = "isbn", how="inner")
```

```python
df_ratings.columns
```

```python


df_ratings = df_ratings.drop_duplicates(['user', 'title'])


pivoted_df=df_ratings.pivot(index='title', columns='user', values='rating')
pivoted_df = pivoted_df.fillna(0)
pivoted_df
```

```python


model = NearestNeighbors(algorithm="brute",metric="cosine", n_jobs=-1)
model.fit(pivoted_df)

```

```python
# add your code here - consider creating a new cell for each section of code
```

```python
# function to return recommended books - this will be tested
def get_recommends(book = ""):
    x = pivoted_df[pivoted_df.index == book]
    dist, ind = model.kneighbors(x, n_neighbors = 6)
    dist, ind = dist.flatten(), ind.flatten()
    recommended_books = []
    recommended_books.append(pivoted_df.index[ind[0]])
    recs=[]
    for i in reversed(range(len(dist))):
        if i !=0:
            recs.append([pivoted_df.index[ind[i]], dist[i]])
        else:
            pass
    recommended_books.append(recs)

    return recommended_books

```

```python
get_recommends("Where the Heart Is (Oprah's Book Club (Paperback))")
```

```python
books = get_recommends("Where the Heart Is (Oprah's Book Club (Paperback))")
print(books)

def test_book_recommendation():
    test_pass = True
    recommends = get_recommends("Where the Heart Is (Oprah's Book Club (Paperback))")
    if recommends[0] != "Where the Heart Is (Oprah's Book Club (Paperback))":
        test_pass = False
    recommended_books = ["I'll Be Seeing You", 'The Weight of Water', 'The Surgeon', 'I Know This Much Is True']
    recommended_books_dist = [0.8, 0.77, 0.77, 0.77]
    for i in range(2): 
        if recommends[1][i][0] not in recommended_books:
            test_pass = False
        if abs(recommends[1][i][1] - recommended_books_dist[i]) >= 0.05:
              test_pass = False
    if test_pass:
        print("You passed the challenge! ðŸŽ‰ðŸŽ‰ðŸŽ‰ðŸŽ‰ðŸŽ‰")
    else:
        print("You haven't passed yet. Keep trying!")

test_book_recommendation()
```
